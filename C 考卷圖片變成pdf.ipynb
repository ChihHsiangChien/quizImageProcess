```python
import csv
import os
import pandas as pd
import numpy as np

from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw

```

```python
df = pd.read_csv('data.csv')

qList = []
for i in range(10):
    qList.append(str(i*5+4))

qList.append('id')
score = df.loc[:,qList]

nameList = df['id']
scoreList = list(np.sum(score,axis = 1))

nameScoreList = list(zip(nameList, scoreList))
```

```python
nameScoreList
df = pd.DataFrame(nameScoreList) 

df.to_csv('總分.csv' ,encoding = 'utf8')
```
