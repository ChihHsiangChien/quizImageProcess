```python
def make_new_student_img(i):
    #取得目錄內所有圖片檔
    path1 = r'G:\我的雲端硬碟\Classroom\阿簡生物7\109學年度第二學期七年級生物科第三次段考 (File responses)'
    
    path2 = img_folder_list[i]

    imgPath = os.path.join(path1,path2)
    for path, subdirs, files in os.walk(imgPath):
        for name in files:
            #print(os.path.join(path, name))
            imgFullPath = os.path.join(path, name)
            print(imgFullPath)
            #print(name)
            name = os.path.splitext(name)
            #print(name[0])
            if ' - 109' in name[0] :
                start =  name[0].index(' - 109')
            if '.' in name[0]:
                end = name[0].index('.')
            #print(name[0][start+3:])
            student_name = name[0][start+3:]
            print(student_name)



            #排除亂交作業的
            if name[1] != '.gdoc'  and name[1] != '.ini'  and name[1] != '.docx':
                answerImg = Image.open(imgFullPath)
                
                width , height = answerImg.size

                ratio = height/width
                
                if ratio > 0.75:
                    newHeight = 600
                    newWidth = int(newHeight / ratio)
                    
                else:
                    newWidth = 800
                    newHeight = int(height/width*newWidth)

                answerImg_resize = answerImg.resize( (newWidth, newHeight) ) 
                
                
                blankImg = Image.open('blankImage/blank'+ str(i) + '.png')

                blankImg.paste(answerImg_resize,(20,50))

                #寫入學生的姓名
                draw = ImageDraw.Draw(blankImg)

                draw.text((850, 50 ), student_name ,(0,0,0),font=font)


                blankImg.save('answerImage/' + str(i) + '/' + student_name + '_'+ str(i) + '.png')


    
```
